```python
import glob
from skimage import io
import matplotlib.pyplot as plt
import cv2
import os
```

```python
images_paths = sorted(glob.glob('images_0725/*.jpg'))
images = [io.imread(p) for p in images_paths]
images_paths
```

```python
plt.imshow(images[6])
```

```python
image = images[6]
image = image[2200:2270, 2860:2970]
plt.imshow(image)
```

```python
output_dir = '/Users/hyunseo.lee/smartphone_capillaroscope/cropped'
os.makedirs(output_dir, exist_ok=True)

for idx, img in enumerate(images):
    cropped = img[2200:2270, 2860:2970]
    out_path = os.path.join(output_dir, f'cropped_{idx:03d}.png')
    cv2.imwrite(out_path, cropped)

print(f"Saved {len(images)} crops to {output_dir}")
```
